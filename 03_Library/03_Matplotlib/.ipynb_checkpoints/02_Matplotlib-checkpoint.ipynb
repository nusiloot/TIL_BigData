````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.arange(0, 10, 1)
y1 = x
y2 = x + 2
y3 = x + 4
y4 = x + 6
````

````python
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
````

````python
plt.plot(x, y1, 'm-', x, y2, 'y:', x, y3, 'k-.', x, y4, 'c')
plt.show()
````

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.show()
````

````python
x = np.arange(-4.5, 5, 0.5)
y = 2*x**3
````

````python
plt.plot(x, y)
plt.title('Graphe title')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)  # 격자 생성 시 .grid(True) 대신에 .grid()라고 써도 됨
plt.show()
````

````python
x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
````

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1', 'data2', 'data3', 'data4'])
plt.show()
````

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.legend(['data1', 'data2', 'data3', 'data4'], loc = 'lower right')
plt.show()
````

````python
plt.plot(x, y1, '>--r', x, y2, 's-g', x, y3, 'd:b', x, y4, '-.Xc')
plt.title('Graph Title')
plt.xlabel('X-value')
plt.ylabel('Y-value')
plt.grid()
plt.legend(['data1', 'data2', 'data3', 'data4'], loc = 'lower right')
plt.show()
````

### cf) matplotlib 라이브러리로 그래프 그릴 때 한글 폰트 사용 방법
#### (참고 사이트 : https://programmers.co.kr/learn/courses/21/lessons/950)

- 그래프 한글 표시법 : matplotlib에서 사용하는 폰트를 한글 폰트로 지정해야 함
```python
import matplotlib
matplotlib.rcParams['font.family']
```
- 폰트 변경법
```python
matplotlib.rcParams['font.family'] = '폰트 이름'
matplotlib.rcParams['axes.unicode_minus'] = False  # 한글 폰트 지정 후 그래프에서 마이너스(-) 폰트가 깨지는 문제 해결
```
- matplotlib에서 사용 가능한 폰트 이름
```python
import matplotlib.font_manager as fm
font_list = fm.findSystemFonts(fontpaths=None, fontext='ttf')
f = [f.name for f in fm.fontManager.ttflist]  # 폰트 이름만 출력
```
- 한글 폰트 사용 방법(1) - FontProperties
```python
path = '~~'  # 해당 폰트의 파일 경로
fontprop = fm.FontProperties(fname=path, size=18)
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.ylabel('주식 가격', fontproperties=fontprop)
plt.xlabel('시간(분)', fontproperties=fontprop)
plt.show()
```
- 한글 폰트 사용 방법(2) - matplotlib.rcParams[]으로 전역 글꼴 설정
```python
plt.rcParams["font.family"] = 'Nanum Brush Script OTF'
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (14,4)
```

````python
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
````
